## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "../Resources/mouse_metadata.csv"
study_results_path = "../Resources/study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_study_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
combined_study_df.head(20)
```

```python
# Checking the number of mice.
combined_study_mice = len(combined_study_df)
print(combined_study_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#pokedex_group
grouped_mice_df = combined_study_df.groupby(["Mouse ID", "Timepoint"]).mean()
grouped_mice_df
grouped_mice_df.head(40)
```

```python
mouse_count = len(combined_study_df["Mouse ID"].unique())
print(mouse_count)
```

```python
duplicate_check = grouped_mice_df[grouped_mice_df.duplicated(keep=False)]
duplicate_check
```

```python
#ins_groupby
clean_grouped_df = duplicate_check.dropna(how="any")
clean_grouped_df.count()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_check.head(20)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
dropped_mice_df = grouped_mice_df.drop_duplicates(keep=False)
dropped_mice_df
```

```python
# Checking the number of mice in the clean DataFrame.
dropped_mice_df.count()
```
